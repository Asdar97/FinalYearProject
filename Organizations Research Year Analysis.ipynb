```python
def organization_year(data):
    a = data['Publication Year']
    b = data['Addresses'].str.upper()
    
    df = pd.DataFrame()
    df = df.assign(Year=a, Organizations=b)
    df = df.dropna()
    #print(len(df))
    
    dfx = df.loc[df['Year'] == 2020]
    dfx = dfx.reset_index(drop=True)
    print(dfx)
    print('Total:',len(dfx))
    
    df1 = dfx[dfx['Organizations'].str.contains(r'UNIV KEBANGSAAN MALAYSIA|\bUKM\b|NATL UNIV MALAYSIA')]
    df1 = df1.reset_index(drop=True)
    #print(df1)
    print('UKM:',len(df1))
    
    df2 = dfx[dfx['Organizations'].str.contains(r'UNIV TEKNOL MALAYSIA|UNIV TECHNOL MALAYSIA|UNIV TEKNOLOGI MALAYSIA|\bUTM\b')]
    df2 = df2.reset_index(drop=True)
    #print(df2)
    print('UTM:',len(df2))
    
    df3 = dfx[dfx['Organizations'].str.contains(r'UNIV MALAYA|UNIV MALAYSIA, KUALA LUMPUR|\bUM\b')]
    df3 = df3.reset_index(drop=True)
    #print(df3)
    print('UM:',len(df3))
    
    df4 = dfx[dfx['Organizations'].str.contains(r'UNIV SAINS MALAYSIA|\bUSM\b')]
    df4 = df4.reset_index(drop=True)
    #print(df4)
    print('USM:',len(df4))
    
    df5 = dfx[dfx['Organizations'].str.contains(r'UNIV MALAYSIA PERLIS|\bUniMAP\b')]
    df5 = df5.reset_index(drop=True)
    #print(df5)
    print('UniMAP:',len(df5))
    
    df6 = dfx[dfx['Organizations'].str.contains(r'UNIV MALAYSIA PAHANG|\bUMP\b')]
    df6 = df6.reset_index(drop=True)
    #print(df6)
    print('UMP:',len(df6))
    
    df7 = dfx[dfx['Organizations'].str.contains(r'UNIV PUTRA MALAYSIA|UNIV PERTANIAN MALAYSIA|\bUPM\b')]
    df7 = df7.reset_index(drop=True)
    #print(df7)
    print('UPM:',len(df7))
    
    df8 = dfx[dfx['Organizations'].str.contains(r'UNIV TEKNOL MARA|UNIV TECHNOL MARA|\bUiTM\b')]
    df8 = df8.reset_index(drop=True)
    #print(df8)
    print('UiTM:',len(df8))
    
    df9 = dfx[dfx['Organizations'].str.contains(r'UNIV TEKNOL PETRONAS|UNIV TECHNOL PETRONAS|\bUTP\b')]
    df9 = df9.reset_index(drop=True)
    #print(df9)
    print('UTP:',len(df9))
    
    df10 = dfx[dfx['Organizations'].str.contains(r'AGENSI NUKLEAR MALAYSIA|MALAYSIAN NUCL AGCY|MALAYSIAN INST NUCL|\bMINT\b')]
    df10 = df10.reset_index(drop=True)
    #print(df10)
    print('MINT:',len(df10))
    
    pd.options.display.max_colwidth = 50
    

    
organization_year(data1)   
```

```python
def research_year(data):
    a = data['Publication Year']
    b = data['Research Areas']
    
    df = pd.DataFrame()
    df = df.assign(Year=a, Research_Areas=b)
    df = df.dropna()
    #print(df)
    
    dfx = df.loc[df['Year'] == 2007]
    dfx = dfx.reset_index(drop=True)
    print(dfx)
    print('Total:',len(dfx))
    
    df1 = dfx[dfx['Research_Areas'].str.contains(r'Energy & Fuels')]
    df1 = df1.reset_index(drop=True)
    #print(df1)
    print('Energy & Fuels:',len(df1))
    
    df2 = dfx[dfx['Research_Areas'].str.contains(r'Chemistry')]
    df2 = df2.reset_index(drop=True)
    #print(df2)
    print('Chemistry:',len(df2))
    
    df3 = dfx[dfx['Research_Areas'].str.contains(r'Electrochemistry')]
    df3 = df3.reset_index(drop=True)
    #print(df3)
    print('Electrochemistry:',len(df3))
    
    df4 = dfx[dfx['Research_Areas'].str.contains(r'Engineering')&\
              ~dfx['Research_Areas'].str.contains(r'Metallurgical Engineering')]
    df4 = df4.reset_index(drop=True)
    print(df4)
    print('Engineering:',len(df4))
    
    df5 = dfx[dfx['Research_Areas'].str.contains(r'Materials Science')]
    df5 = df5.reset_index(drop=True)
    #print(df5)
    print('Materials Science:',len(df5))
    
    df6 = dfx[dfx['Research_Areas'].str.contains(r'Science & Technology - Other Topics')]
    df6 = df6.reset_index(drop=True)
    #print(df6)
    print('Science & Technology - Other Topics:',len(df6))
    
    df7 = dfx[dfx['Research_Areas'].str.contains(r'Environmental Sciences & Ecology')]
    df7 = df7.reset_index(drop=True)
    #print(df7)
    print('Environmental Sciences & Ecology:',len(df7))
    
    df8 = dfx[dfx['Research_Areas'].str.contains(r'Polymer Science')]
    df8 = df8.reset_index(drop=True)
    #print(df8)
    print('Polymer Science:',len(df8))
    
    df9 = dfx[dfx['Research_Areas'].str.contains(r'Physics')]
    df9 = df9.reset_index(drop=True)
    #print(df9)
    print('Physics:',len(df9))
    
    df10 = dfx[dfx['Research_Areas'].str.contains(r'Biotechnology & Applied Microbiology')]
    df10 = df10.reset_index(drop=True)
    #print(df10)
    print('Biotechnology & Applied Microbiology:',len(df10))
    
    pd.options.display.max_colwidth = 200

research_year(data1)   
```
