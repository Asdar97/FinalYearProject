```python
def Adresses(x): 
    [str(x) for x in x['Addresses']] 
    x1 = x['Addresses'].str.replace(r"(\s*\[.*?\]\s*)", " ").str.strip() 
    #x1 = x1['Addresses'].str.replace(r"(\s*\(.*?\)\s*)", " ").str.strip() 
    new = x1.str.split('$ ', expand = True)

    #print(new)
    fulldata = pd.DataFrame()
    i = 0
    j = (len(new.columns)-1)
    while i <= j:
        data = pd.DataFrame()
        data['Addresses']= new[i]
        data = data.dropna()
        #print(data)
        fulldata = fulldata.append(data)
        fulldata = fulldata.reset_index(drop=True)
        i+=1

    fulldata = fulldata['Addresses'].str.upper()
    #print(fulldata)
    zz = len(fulldata)
    print(zz)

    
    #pd.options.display.max_colwidth = 500
    fulldata = pd.DataFrame(fulldata, columns= ['Addresses'])
    df1 = fulldata[fulldata['Addresses'].str.contains(r'AGENSI NUKLEAR MALAYSIA|MALAYSIAN NUCL AGCY|MALAYSIAN INST NUCL|\bMINT\b')]
    df1 = df1.reset_index(drop=True)
    #print(df1)
    yy = len(df1)
    print(yy)

    df2 = df1[~df1['Addresses'].str.contains(r'\bMINT\b')]
    df2 = df2.reset_index(drop=True)
    #print(df2)
    xx = len(df2)
    print(xx)

Adresses(data1)
```
