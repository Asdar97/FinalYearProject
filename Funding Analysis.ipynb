```python
def funding(x):
    a = x['Addresses'].str.upper()
    b = x['Article Title'].str.lower()
    c = x['Publication Year']
    d = x['Funding Orgs'].str.lower()
    e = x['Funding Text'].str.lower()
    f = x['Times Cited, WoS Core']
    
    df = pd.DataFrame()
    df = df.assign(Organizations=a, Title=b, Publication_Year=c, Funding_Orgs=d, Funding_Text=e, Citations=f)
    df = df.reset_index(drop=True)
    #print(df)
    
    df1 = df[df['Organizations'].str.contains(r'UNIV TEKNOL MALAYSIA|UNIV TECHNOL MALAYSIA|UNIV TEKNOLOGI MALAYSIA|\bUTM\b')]
    df1 = df1.reset_index(drop=True)
    #print(df1,'\n')
    
    df1 = df1[df1['Publication_Year'] == 2009]
    df1 = df1.reset_index(drop=True)
    df1.dropna(subset = ["Funding_Orgs"], inplace=True)
    df1 = df1.sort_values(by=['Citations'], ascending=False)
    df1 = df1.reset_index(drop=True)
    #print(df1,'\n')
    
    #df1.to_csv('Funding Organizations (UTM) - 2009.csv', index=False)
    
    df2 = df1[df1['Funding_Orgs'].str.contains(r'universiti kebangsaan malaysia')]
    df2 = df2.reset_index(drop=True)
    print('Universiti Kebangsaan Malaysia:', len(df2))
    
    df8 = df1[df1['Funding_Orgs'].str.contains(r'ministry of higher education')]
    df8 = df8.reset_index(drop=True)
    print('Ministry of Higher Education (MOHE):', len(df8))
    
    df9 = df1[df1['Funding_Orgs'].str.contains(r'ministry of education')]
    df9 = df9.reset_index(drop=True)
    print('Ministry of Education Malaysia:', len(df9))

    
funding(data1) 
```
