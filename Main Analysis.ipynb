```python
def f_total(x,y,z):
    X = x.shape[0]
    Y = y.sum()
    Z = z.sum()
    n = Y/X
    N = round(n,2)
    a = sum(x >= i + 1 for i, x in enumerate(sorted(  list(y), reverse=True)))
    print('Total Publications:',X)
    print('Total citations in WoS core:',Y)
    #print('Times cited in all data base:',Z)
    print('Total citations per total publications:',n,'≈',N)
    print('h-index=',a)
```

```python
def f_bar(x):
    a = x.iloc[:,44]
    b = x.iloc[:,31]
    table = pd.DataFrame()
    table = table.assign(Publication_Year=a,times_cited=b)
    table['Total'] = table.groupby(['Publication_Year'])['times_cited'].transform('sum')
    table = table.drop_duplicates(subset=['Publication_Year'])
    table = table.dropna()
    d = table.iloc[:,0]
    f = table.iloc[:,2]
    e = x.pivot_table(index=['Publication Year'], aggfunc='size')
    e = pd.DataFrame(e, columns= ['Publications'])
    listy= e['Publications'].values.tolist()
    table_2 = pd.DataFrame()
    table_2 = table_2.assign(Publication_Year=d,total_citations=f)
    table_2 = table_2.sort_values('Publication_Year')
    table_2 = table_2.assign(Publications=listy)
    table_2 = table_2.reset_index(drop=True)
    g = table_2.iloc[:,1]/table_2.iloc[:,2]
    g = round(g,2)
    X = x.shape[0]
    h = (table_2.iloc[:,2]/X)*100
    h = round(h,2)
    ratio = []
    i = 0
    j = (table_2.shape[0]-2)
    while i <= j:
        ab = table_2['Publication_Year'].iloc[-1]-table_2['Publication_Year'].iloc[i]
        ac = table_2['total_citations'].iloc[i]/ab
        ac = round(ac,2)
        ratio.append(ac)
        i+=1
    k = table_2['total_citations'].iloc[-1]
    ratio.append(k)
    ba = table_2.iloc[:,1].sum()
    be = (table_2.iloc[:,1]/ba)*100
    be = round(be,2)
    table_2 = table_2.assign(ratio_P_to_TP=h,C_per_TC=be,TC_Per_Y=ratio,TC_per_P=g,)
    
    columnsTitles=['Publication_Year','Publications','ratio_P_to_TP','total_citations','TC_per_P','C_per_TC','TC_Per_Y']
    table_2=table_2.reindex(columns=columnsTitles)
    print(table_2)
    
    table_2.to_csv('publication yearAI.csv', index=False)
    
    bc = table_2['Publication_Year'].iloc[-1]-table_2['Publication_Year'].iloc[0]
    bd = ba/bc
    bd = round(bd,2)
    print('\nAverage total citations per year:',bd)
```

```python
def f_main(data):
    f_total((data.iloc[:,1]),(data.iloc[:,31]),(data.iloc[:,32]))
    f_bar(data)
```

```python
def Research_Area_2(x):
    [str(x) for x in x['Research Areas']]
    new = x['Research Areas'].str.split('; ', expand = True)
    #print(new)
    fulldata = pd.DataFrame()
    i = 0
    j = (len(new.columns)-1)
    while i <= j:
        data = pd.DataFrame()
        data["Research Areas_1"]= new[i]
        a = x.iloc[:,31]
        data = data.assign(total_cited=a)
        data = data.dropna()
        data = data.reset_index(drop=True)
        #print(data)
        fulldata = fulldata.append(data)
        fulldata = fulldata.reset_index(drop=True)
        i+=1
    
    fulldata['Total'] = fulldata.groupby(['Research Areas_1'])['total_cited'].transform('sum')
    fulldata = fulldata.sort_values(by=['Research Areas_1'], ascending=True)
    fulldata = fulldata.drop_duplicates(subset=['Research Areas_1'])
    fulldata = fulldata.reset_index(drop=True)
    #print(fulldata)
    
    return(fulldata.iloc[:,2])
```

```python
def Research_Area(x):
    #listx= x['Research Areas'].values.tolist()
    listx= x.iloc[:,59].values.tolist()
    results = []
    for i in listx:
        y = i.split(';')
        j=0
        k = len(y)-1
        while j <= k:
            results.append(y[j])
            j+=1
    
    results=[i.strip(' ') for i in results]
    
    c = x.iloc[:,31]
    ba = c.sum()
    
    df = pd.DataFrame(results, columns=['Research Area'])
    e = df.pivot_table(index=['Research Area'], aggfunc='size')
    e = pd.DataFrame(e, columns= ['Publications'])
    e = e.reset_index().rename_axis(None, axis=1)
    e = e.sort_values(by=['Research Area'], ascending=True)
    e = e.reset_index(drop=True)
    ab = Research_Area_2(data1)
    X = x.shape[0]
    b = (e.iloc[:,1]/X)*100
    b = round(b,2)
    bb = round((ab/ba)*100,2)
    e = e.assign(ratio_P_to_TP=b,total_citations=ab, ratio_C_to_TC=bb)
    ac = e.iloc[:,3]/e.iloc[:,1]
    ac = round(ac,2)
    e = e.assign(TC_per_P=ac)
    e = e.sort_values(by=['Publications'], ascending=False)
    e = e.reset_index(drop=True)
    print(e.head(15))
    
    #e.to_csv('research areaAI.csv', index=False)
    
    ab = e.shape[0]
    print('\nTotal Research Area:',ab)
    
Research_Area(data1)
```

```python
def Authors(x):
    #listx= x['Author Full Names'].values.tolist()
    listx= x.iloc[:,1].values.tolist()
    results = []
    for i in listx:
        y = i.split(';')
        j=0
        k = len(y)-1
        while j <= k:
            results.append(y[j])
            j+=1
    
    results=[i.strip(' ') for i in results]
    
    df = pd.DataFrame(results, columns=['Authors'])
    e = df.pivot_table(index=['Authors'], aggfunc='size')
    e = pd.DataFrame(e, columns= ['Publications'])
    e = e.reset_index().rename_axis(None, axis=1)
    e = e.sort_values(by=['Authors'], ascending=True)
    e = e.reset_index(drop=True)
    ab = Authors_2(x)
    X = x.shape[0]
    b = (e.iloc[:,1]/X)*100
    b = round(b,2)
    ba = x['Times Cited, WoS Core'].sum()
    bb = round((ab/ba)*100,2)
    e = e.assign(ratio_P_to_TP=b,total_citations=ab, ratio_C_to_TC=bb)
    ac = e.iloc[:,3]/e.iloc[:,1]
    ac = round(ac,2)
    e = e.assign(TC_per_P=ac)
    e = e.sort_values(by=['Publications'], ascending=False)
    e = e.reset_index(drop=True)
    print(e.head(15))
    
    #e.to_csv('AuthorsAI.csv', index=False)
    
    ab = e.shape[0]
    print('\nTotal Authors:',ab)
    
Authors(data1)
```

```python
def Journals(x):
    a = x.pivot_table(index=['Source Title'], aggfunc='size')
    a = pd.DataFrame(a, columns= ['Publications'])
    a = a.reset_index().rename_axis(None, axis=1)
    a = a.sort_values(by=['Source Title'], ascending=True)
    a = a.reset_index(drop=True)
    
    b = x.iloc[:,9]
    c = x.iloc[:,31]
    ab = c.sum()
    
    table = pd.DataFrame()
    table = table.assign(Journals=b,times_cited=c)
    table['Total'] = table.groupby(['Journals'])['times_cited'].transform('sum')
    table = table.drop_duplicates(subset=['Journals'])
    table = table.dropna()
    table = table.sort_values(by=['Journals'], ascending=True)
    table = table.reset_index(drop=True)
    d = table.iloc[:,2]
    f = x.shape[0]
    g = (a.iloc[:,1]/f)*100
    g = round(g,2)
    h = table.iloc[:,2]/a.iloc[:,1]
    h = round(h,2)
    ac = round((d/ab)*100,2)
    e = a.assign(total_citations=d, ratio_P_per_TP=g, TC_per_P=h, ratio_C_per_TC=ac)
    e = e.sort_values(by=['Publications'], ascending=False)
    e = e.reset_index(drop=True)
    print(e.head(15))
    
    #e.to_csv('JournalsAI.csv', index=False)
    
    ad = e.shape[0]
    print('\nTotal Journals:',ad)

Journals(data1)
```

```python
def Articles(x):
    a = x.iloc[:,8]
    b = x.iloc[:,1]
    c = x.iloc[:,44]
    d = x.iloc[:,9]
    e = x['Times Cited, WoS Core']
    k = x['Research Areas']
    g = x['Addresses']
    table = pd.DataFrame()
    table = table.assign(Title=a,Authors=b,Year=c,Research_Areas=k, Journal=d, Citations=e)

    ratio = []
    k = table.iloc[:,2]
    for i in k:
        if i == 2020.0:
            ab = 1
            ratio.append(ab)
        else:
            ab = 2020-i
            ratio.append(ab)
            
    df = pd.DataFrame(ratio)
    f = table.iloc[:,5]/df.iloc[:,0]
    f = round(f,2)   
    table = table.assign(citations_per_year=f, Addresses=g,)
    table = table.sort_values(by=['Citations'], ascending=False)
    table = table.reset_index(drop=True)
    print(table.head(15))
    
    table.to_csv('ArticlesAI.csv', index=False)
    
    ac = table.shape[0]
    print('\nTotal Articles:',ac)
    
    #articles = ExcelWriter('Articles.xlsx')
    #table.to_excel(articles, index=False)
    #articles.save()
    
    #table.head(20).to_csv('TOP_Articles.csv')
Articles(data1)
```
